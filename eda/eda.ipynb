# **3. Обработка данных**  

````python
# 1. Анализ пропущенных значений
missing_values = df.isnull().sum()
missing_percent = (missing_values / df.shape[0]) * 100
missing_data = missing_percent[missing_percent > 0].sort_values(ascending=False).round(2)
missing_data
````

````python
# Замена пропущенных значений
if not missing_data.empty:
    for col, percent in missing_data.items():
        if percent > 60:
            df.drop(columns=[col], inplace=True)
        elif percent > 30:
            df[col].fillna(df[col].median(), inplace=True)
        elif percent > 5:
            df[col].fillna(df[col].mode()[0], inplace=True)
        else:
            df.dropna(subset=[col], inplace=True)
````

````python
# Повторный просмотр пропущенных значений
missing_values = df.isnull().sum()
missing_percent = (missing_values / df.shape[0]) * 100
missing_data = missing_percent[missing_percent > 0].sort_values(ascending=False).round(2)
missing_data
````

````python
# Проверка дубликатов
duplicate_rows = df.duplicated().sum()
if duplicate_rows > 0:
    print("Дубликаты были удалены")
    df.drop_duplicates(inplace=True)
else:
    print("В данных отсутствуют дубликаты")
````

````python
# Выбросы (по IQR)
numerical_cols = df.select_dtypes(include="number").columns

outlier_decisions = {}
outliers_dict = {}

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_count = outliers.shape[0]
    
    if outlier_count > 0:
        outliers_dict[col] = outlier_count

        if outlier_count / len(df) > 0.1: 
            df[col] = df[col].clip(lower=df[col].quantile(0.05), upper=df[col].quantile(0.95))
            outlier_decisions[col] = "Ограничены 5%-95% квантилями"
        elif outlier_count / len(df) < 0.02: 
            df = df[~((df[col] < lower_bound) | (df[col] > upper_bound))]
            outlier_decisions[col] = "Удалены, так как их количество незначительно"
        else:
            df[col] = df[col].mask((df[col] < lower_bound) | (df[col] > upper_bound), df[col].mean())
            outlier_decisions[col] = "Заменены средним значением"
            
outlier_decisions = pd.Series(outlier_decisions)
outlier_decisions
````
